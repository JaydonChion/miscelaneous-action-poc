```python
import numpy as np
import glob
import cv2
d={}
```

```python
data_dir='editedut/*/*'
data_list=glob.glob(data_dir)
img_size=(28,28)
frame_number=20
data=np.empty([1,frame_number,img_size[0],img_size[1],3])
label=np.empty([1,])
fps=30
sample_step=1
npz_order=0
num_cat = 2
video_per_cat=14


print(len(data_list))
print(data.shape)
```

```python
with open("classIndVio.txt") as f:
    for line in f:
       (key, val) = line.split()
       d[int(key)-1] = val
```

```python
one_hot_array = np.zeros((1,num_cat))
```

```python
videonumkeeptrack = np.zeros((1,num_cat))
```

```python
arraynum = 0
for num, video in enumerate(data_list):
    read_video = False
    print (num, video)

    for key, videoname in d.items():    # for name, age in dictionary.iteritems():  (for Python 2.x)
        if (videoname in video):
            cap = cv2.VideoCapture(video)
            #fps = cap.get(cv2.CAP_PROP_FPS)
            total_frame=cap.get(cv2.CAP_PROP_FRAME_COUNT)
#             if(videonumkeeptrack[0][key]<video_per_cat):
#                 videonumkeeptrack[0][key] = videonumkeeptrack[0][key]+1
#                 print(videonumkeeptrack)
#                 print(video)
#                 one_hot_array[arraynum][key]=1

            sample=np.zeros([1,1,img_size[0],img_size[1],3])
            for i in range(int(total_frame)):
                if (i%sample_step)==0 and not i == 0:
                    ret, frame = cap.read()
                    if ret==False:
                        prev=sample[0,-1].reshape(1,1,img_size[0],img_size[1],3)
                        sample=np.append(sample,prev,axis=1)
                    else:
                        cropped_frame=cv2.resize(frame,(img_size[0],img_size[1]))
                        cropped_frame=cropped_frame.reshape(1,1,img_size[0],img_size[1],3)
                        sample=np.append(sample,cropped_frame,axis=1)
                        print(sample.shape)
                    if((i%(frame_number) )== 0 and (not i == 0)):
                        print("trim")
                        print(sample.shape)

                        sample = np.moveaxis(sample,1,0)
                        sample=sample[1:]
                        sample = np.moveaxis(sample,1,0)
                        print("the shape is")
                        print(sample.shape)

                        sample=sample.reshape([1,frame_number,img_size[0],img_size[1],3])
#                         #print("one sample shape: {}".format(sample.shape))
                        print(sample.shape)
                        data=np.append(data,sample,axis=0)
                        sample=np.zeros([1,1,img_size[0],img_size[1],3])

                        if("violence" in video):
                            one_hot_array = np.append(one_hot_array,[[1,0]],axis=0)
                        elif("non" in video):
                            one_hot_array = np.append(one_hot_array,[[0,1]],axis =0)
                        

#         del sample,cropped_frame

one_hot_array = one_hot_array[1:,...]
print("label shape{}".format(one_hot_array.shape))
print("data shape{}".format(data.shape))
```
