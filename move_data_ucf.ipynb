```python
import os
import glob
from shutil import copyfile
import shutil
data_dir='../../Dataset_AR/UCF101/*'
target_dir="../../Dataset_AR/ucf/"
data_list=glob.glob(data_dir)
print(data_list[0:5])
class_num=2
video_num=0
```

```python
def exam_name(name):
    if "_Swing" in name or "_BrushingTeeth" in name or "_GolfSwing" in name:
        return True
    else:
        return False
```

```python
def process_name(name):
    name=name.split("/")[-1]
    name=name.split("_")[1]
    return name
```

```python
if os.path.exists('../../Dataset_AR/ucf'):
    shutil.rmtree('../../Dataset_AR/ucf')
    os.makedirs('../../Dataset_AR/ucf')
else:    
    os.makedirs('../../Dataset_AR/ucf')
    
ind=0
for index,video in enumerate(data_list):
    bare_name=process_name(video)
    status=exam_name(video)
    path=video
    if status:
        if not os.path.exists('../../Dataset_AR/ucf/{}'.format(bare_name)):
            os.makedirs('../../Dataset_AR/ucf/{}'.format(bare_name))
        print(bare_name,ind)
        copyfile(video, "../../Dataset_AR/ucf/{}/{}_{}".format(bare_name,bare_name,ind))
        ind+=1
```
