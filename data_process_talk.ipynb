```python
import numpy as np
import glob
import cv2
```

```python
data_dir='talkdata/*/*'
data_list=glob.glob(data_dir)
img_size=(28,28)
frame_number=20
sample_step=3
data=np.empty([1,frame_number,img_size[0],img_size[1],3])
npz_order=0
print(len(data_list))
print(data.shape)
```
